```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv(r'C:\data\dec_21\train.csv')
train_data.head()
```

```python
test_data = pd.read_csv(r'C:\data\spaceship-titanic\test.csv')
```

```python
print('{0} number of training entries'.format(len(train_data)))
print('{0} number of test entries'.format(len(test_data)))
```

```python
col = 'Cover_Type'

X = pd.get_dummies(train_data.loc[:, train_data.columns != col])
X_test = pd.get_dummies(test_data.loc[:, test_data.columns != col])

y = pd.get_dummies(train_data[col])
```

```python
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(7, activation='softmax')
])

model.compile(
    loss=tf.keras.losses.categorical_crossentropy,
    optimizer=tf.keras.optimizers.Adam(lr=0.03),
    metrics=[
        tf.keras.metrics.BinaryAccuracy(name='accuracy'),
        tf.keras.metrics.Precision(name='precision'),
        tf.keras.metrics.Recall(name='recall')
    ]
)

history = model.fit(X, y, epochs=100)
```
